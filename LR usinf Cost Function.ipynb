```python
x = [1,2,3,4,5]
y = [1,4,6,8,10]

w = 0
b = 0

N = len(x)
alpha = 0.01
epoch = 1000

for i in range(epoch):
    y_pred = [w * x[j] + b for j in range(N)]
    error = [y[j] - y_pred[j] for j in range(N)]

    djw_ = [(-x[j]*(y[j]-y_pred[j]))/N for j in range(N)]
    djb_ = [(-(y[j] - y_pred[j]))/N for j in range(N)]

    djw = sum(djw_)
    djb = sum(djb_)

    w = w - alpha*djw
    b = b - alpha*djb

print(w,b)
```
